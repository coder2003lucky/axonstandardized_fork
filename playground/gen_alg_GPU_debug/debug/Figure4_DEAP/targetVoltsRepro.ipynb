```python
def nrnMreadH5(fileName):
    f = h5py.File(fileName,'r')
    dat = f['Data'][:][0]
    return np.array(dat)


curr_volts = nrnMreadH5("../Data/VHotP0.h5")
Nt = int(len(curr_volts)/5)
shaped_volts = np.reshape(curr_volts, [5, Nt])
plt.plot(shaped_volts[0])
```
