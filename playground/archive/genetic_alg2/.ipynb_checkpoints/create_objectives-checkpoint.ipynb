```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import os
import scipy.stats as stat
import sys
```

```python
single_stim_opt = h5py.File('../analyze_p/optimization_results_path/opt_result_single_stim_bbp_full.hdf5')
opt_file_multistims = h5py.File('./objectives/multi_stim_bbp_full.hdf5')
stims_optimal_order = [e.decode('ascii') for e in single_stim_opt['stims_optimal_order'][:]]
ordered_score_function_list = [e.decode('ascii') for e in opt_file_multistims['ordered_score_function_list'][:]]
score_dir = '../scores/'
```

```python
# Search for best sf using spearman
sf_name_dict = {}
for i in range(len(stims_optimal_order)):
    score_data = h5py.File(scores_dir+stims_optimal_order[i]+'_scores.hdf5', 'r')
    for j in range(len(ordered_score_function_list)):
        curr_score = score_data['norm_pin_scores_'+ordered_score_function_list[j]][:]
        spearman = stat.spearmanr(curr_score, np.array(range(len(curr_score))).reshape((len(curr_score), 1)))[0]
        if np.isnan(spearman):
            spearman = 0
        sf_name_dict[(stims_optimal_order[i], ordered_score_function_list[j])] = spearman
sorted_scores = sorted(sf_name_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
score_len = 10000
for i in range(5):
    stim, sf = sorted_scores[i][0]
    score_data = h5py.File(scores_dir+stim+'_scores.hdf5', 'r')['norm_pin_scores_'+sf[:]]
    plt.figure(figsize=(10, 7))
    plt.scatter(range(1, score_len+1), score_data)
    plt.show()
```

```python
print(sorted_scores[:5])
best_stim, best_sf = sorted_scores[0][0]
weight_list = [1]
```

```python
best_pair = h5py.File('./objectives/best_pair_bbp_full.hdf5', 'w')
best_pair.create_dataset('stims_optimal_order', data=np.array([np.string_(best_stim)]))
best_pair.create_dataset('ordered_score_function_list', data=np.array([np.string_(best_sf)]))
best_pair.create_dataset('optimal_weight_list', data=np.array(weight_list)
```
