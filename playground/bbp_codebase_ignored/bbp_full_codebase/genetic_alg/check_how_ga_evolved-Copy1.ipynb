```python
import numpy as np
import h5py 
import pickle as pkl
import matplotlib.pyplot as plt 
import os
import bluepyopt
```

```python
GA_results = pkl.load(open('./best_indv_logs/best_indvs_gen_99.pkl', 'rb'))
```

```python
GA_results = np.array(GA_results)
```

```python
orig_params = h5py.File('../run_volts_bbp/params/params_bbp_full.hdf5', 'r')['orig_full'][:]
```

```python
active_list = [1,2,3,4,5,6,8,9,10,11,12,14,16,17,18,19,20,22,23]
orig_params = [orig_params[0][e] for e in active_list]
```

```python
deviations = orig_params - GA_results
```

```python
fig = plt.figure(figsize=(15, 15))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for ind in range(deviations.T.shape[0]):
    ax = fig.add_subplot(5, 4, ind+1)
    ax.plot(np.abs(deviations.T[ind])/orig_params[ind])
    ax.set_xlabel('Generations')
    ax.set_ylabel('Mean absolute percent error')
plt.show()
```
